### Preprocessing

```python
# Two functions to fill missing data: DataFrame.fillna(...) & DataFrame.interpolate(...)
DF[DF['station']=='Aotizhongxin'] = DF[DF['station']=='Aotizhongxin'].interpolate(method='linear', axis=0)
DF['PM2.5'] = DF['PM2.5'].interpolate(method='linear', axis=0)
```

```python
import numpy as np
from sklearn.preprocessing import MinMaxScaler

def Normalize(df):
  scaler = MinMaxScaler() ; df_normal = pd.DataFrame(scaler.fit_transform(df), columns=df.columns)
  target_scaler = MinMaxScaler() ; PM_scaled = target_scaler.fit_transform(np.array(df['PM2.5'][:35064]).reshape(-1,1))
  return df_normal, target_scaler, PM_scaled

DF_numerics = DF.select_dtypes(include=np.number)
DF_normal, Target_scaler, PM_scaled = Normalize(DF_numerics) ; DF_normal['station'] = DF.station
DF_normal
```

### Model Functions

```python
# from keras.layers import Input, Conv1D, LSTM, Dense, Activation, BatchNormalization, Dropout, GlobalAveragePooling1D, MaxPooling1D, Embedding
# from sklearn.model_selection import train_test_split
# from keras.models import Sequential
# from keras.optimizers import Adam
# import keras

from tensorflow.keras.layers import Input, Conv1D, LSTM, Dense, Activation, BatchNormalization, Dropout, GlobalAveragePooling1D, MaxPooling1D, Embedding
from sklearn.model_selection import train_test_split
from tensorflow.keras.models import Sequential
from tensorflow.keras.optimizers import Adam
import tensorflow.keras
import keras

def Supervise(dataset, lag): # lag in hours!
  i = 0 ; dataset_lag = np.zeros([int(dataset.shape[0]-lag),int(dataset.shape[1]*lag)]) ; target_lag = np.zeros([int(dataset.shape[0]-lag),1])
  while (i<=dataset.shape[0]-lag-1):
    dataset_lag[i,:] = np.ndarray.flatten(np.array(dataset.loc[i:i+lag-1]))
    target_lag[i,0] = dataset['PM2.5_Aotizhongxin'][i+lag]
    i+=1
  dataset_lag = pd.DataFrame(dataset_lag) ; # target_lag = pd.DataFrame(target_lag)
  X_train, X_test, Y_train, Y_test = train_test_split(dataset_lag, target_lag, test_size=0.2, shuffle = True, random_state = 0)
  X_train = np.expand_dims(X_train, axis=-1) ; X_test = np.expand_dims(X_test, axis=-1)
  return X_train, Y_train, X_test, Y_test

def CNN_LSTM(X_train, Y_train, X_test):
  keras.backend.clear_session()
  model = Sequential()
  model.add(Input(shape = X_train.shape[1:]))
  model.add(Conv1D(64, 3, padding='causal'))
  model.add(Activation('relu'))
  model.add(BatchNormalization())
  model.add(Conv1D(64, 3, padding='causal'))
  model.add(Activation('relu'))
  model.add(BatchNormalization())
  model.add(Conv1D(32, 3, padding='causal'))
  model.add(Activation('relu'))
  model.add(MaxPooling1D(3))
  model.add(LSTM(100, return_sequences=True))
  model.add(Dropout(0.2))
  model.add(LSTM(50))
  model.add(Dropout(0.3))
  model.add(Dense(1))
  model.add(Activation('relu'))
  model.summary()
  ###
  model.compile(optimizer=Adam(learning_rate=0.001,decay=0.0001), loss='mae')
  history = model.fit(X_train, Y_train, epochs=200, batch_size=32, validation_split=0.2, verbose=1, shuffle=False)
  Y_pred = model.predict(X_test)
  ###
  print('')
  plt.figure(figsize=(6,4))
  plt.plot(history.epoch,history.history['loss'])
  plt.plot(history.epoch,history.history['val_loss'])
  plt.grid() ; plt.title('Loss') ; plt.xlabel('Epochs') ; plt.ylabel('Loss') ; plt.legend(['Train', 'Validation']) ; plt.show()
  return Y_pred

def InvScaler(Y):
  return Target_scaler.inverse_transform(Y)

def Asses(Y_test, Y_pred):
  # Y_test = np.expand_dims(Y_test, axis=1)
  print('MAE = ' + str(np.mean(np.abs(Y_test - Y_pred))))
  print('RMSE = ' + str(np.sqrt(np.mean(np.square(Y_test - Y_pred)))))
  print('R2 = ' + str(np.sum(np.square(Y_test - Y_pred))/np.sum(np.square(Y_test - np.mean(Y_test)))))
```

### 1-day Lag

```python
X_train, Y_train, X_test, Y_test = Supervise(dataset, lag=1*24)
Y_pred = CNN_LSTM(X_train, Y_train, X_test)
```

```python
Y_test_denorm = InvScaler(Y_test) ; Y_pred_denorm = InvScaler(Y_pred)
print('Efficiency (normalized outputs with less differences):') ; Asses(Y_test, Y_pred)
print('\nEfficiency (denormalized to the scale of P2.5_Aotizhongxin):') ; Asses(Y_test_denorm, Y_pred_denorm)
```
